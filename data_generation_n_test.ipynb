# DATA GENERATION WITH FAKER

```python
# import required libraries
from faker import Faker
import pandas as pd
```

```python
f = Faker(['tw_GH'])
```

```python
# Create a DataFrame to store the records
header = ["company", "first_name", "last_name", "phone_number", "email_address", "address", "branch", "services_products", "transaction_activity", "transaction_cost", "customer_pref", "comm_method"]
neuce_data = []
```

```python
# define entries to be generated in data and generate data
for _ in range(100000):
        company = f.company()
        first_name = f.first_name()
        last_name = f.last_name()
        phone_number = f.phone_number()
        email = f.email()
        address = f.address()
        branch = f.random_element(elements=("Accra", "Tema", "Kumasi", "Techiman", "Takoradi", "Sunyani"))
        services_products = f.random_element(elements=("Decorative paints", "Waterproofing systems", "Etics/Technical mortars", "Industrial paints", "Flooring", "Road markings"))
        transaction_activity = f.random_element(elements=("Wholesale", "Retail"))
        transaction_cost = f.random_int(min=500, max=10000)
        customer_pref = f.random_element(elements=("Company website", "LinkedIn", "Facebook"))
        comm_method = f.random_element(elements=("Email", "Phone", "Inperson"))

        neuce_data.append([company, first_name, last_name, phone_number, email, address, branch, services_products, transaction_activity, transaction_cost, customer_pref, comm_method])

df = pd.DataFrame(neuce_data, columns=header)
```

```python
# Save the data to a CSV file
df.to_csv("neuce_data.csv", index=False)
```
